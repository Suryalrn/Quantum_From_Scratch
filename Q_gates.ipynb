```python
import numpy as np
class QuantumState:
    def __init__(self,vector):
        self.vector  = np.array(vector)
        self.n = self.vector.size

    def check_power(self):
        if (self.n & (self.n-1)) == 0 and self.n > 0:
            return True
        else:
            return False
            
    def size_qubit(self):
        i =1
        if self.check_power():
            while(self.n!=2**i):
                i+=1
            return i
        else:
            return "Invalid Length of vector"    
        
    def __str__(self):
        L = []
        s = self.size_qubit()
        for j , amp in enumerate(self.vector):    
            if amp != 0:
                binary = format(j,f"0{s}b")
                L.append(f"{amp}|{binary}⟩")
                    
        return(" + ".join(L))
```
