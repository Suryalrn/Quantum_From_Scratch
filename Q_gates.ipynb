```python
import numpy as np

vector  = np.array(list(map(float,input("Enter vector").split())))
n = vector.size
    

def check_power(n):
    if (n & (n-1)) == 0 and n > 0:
        return True
    else:
        return False

def size_qubit(n):
    i =1
    if check_power(n):
        while(n!=2**i):
            i+=1
        return i
    else:
        return "Invalid Length of vector"
s = size_qubit(n)
for j , amp in enumerate(vector):    
    if amp == 0:
        pass
    else:
        binary = format(j,f"0{s}b")
        print(f"|{binary}>")
```
